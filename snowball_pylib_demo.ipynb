## I. Pandas

```python
# import the Pandas library
import pandas as pd
```

```python
# read in dataset as Pandas dataframe
df_cereal = pd.read_csv('cereal.csv')
df_cereal
```

```python
# map column to new values
new_mfr = {"A" : "American Home Food Products",
      "G" : "General Mills",
      "K" : "Kellogs",
      "N" : "Nabisco",
      "P" : "Post",
      "Q" : "Quaker Oatas",
      "R" : "Ralston Purina"}
df_cereal["mfr"] = df_cereal["mfr"].map(new_mfr)
df_cereal
```

```python
# drop unnecessary columns
df_cereal.drop(['shelf', 'weight', 'cups'], axis=1, inplace=True)
df_cereal
```

```python
# slice dataframe to create a smaller dataframe
df_nutrition = df_cereal[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']]
df_nutrition
```

## II. Numpy + Scipy

```python
# import numpy and scipy libraries
import numpy as np
from scipy import stats
```

## III. Matplotlib + Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# distribution plots to identify skew
fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(20, 20))
for i, ax in zip(range(len(df_nutrition.columns)), axes.flat):
    sns.distplot(df_nutrition[df_nutrition.columns[i]], ax=ax, label=df_nutrition.columns[i])
plt.show()
```

```python
# plot cereals
sns.pairplot(df_cereal[['type', 'calories', 'protein', 'fat', 'carbo', 'sugars', 'rating']], hue='type')
```

```python
# heatmap function
def heatmap(dataframe, start_col, end_col, color):
    df_features = dataframe.iloc[:, start_col:end_col]
    
    # find the correlation coefficient between each trait as a new dataframe (absolute value applied so all coefficients between 0 and 1)
    df_corr = df_features.corr().abs()
    
    # set sizes of figure
    fig_dims = (10, 7)
    fig, ax = plt.subplots(figsize=fig_dims)
    
    # create a blue heatmap of strengh of correlation
    viz = sns.heatmap(df_corr, vmin=0, vmax=1, linewidths=0.5, cmap=color, ax=ax)
    viz.set(title="Correlation between Variables")

heatmap(df_cereal, 3, 15, 'Blues')
```

```python
# create pie plot using matplotlib
fig, ax = plt.subplots(figsize=(15,15))
plt.pie(df_cereal['mfr'].value_counts(), labels=df_cereal['mfr'].unique(), autopct='%1.1f%%')
plt.title('Cereal by Manufacturer')
```

```python
# create distribution plot (violinplot) for ratings per manufacturer
fig, ax = plt.subplots(figsize=(15,10))
sns.violinplot(x="mfr", y="rating", data = df_cereal, alpha=.5, ax=ax, palette='Reds')
plt.title('Rating Distribution by Manufacturer')
```

```python
# create bar plot for average sugar levels by manufacturer
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(x="mfr", y="sugars", data = df_cereal, alpha=.5, ax=ax, palette='Purples')
plt.title('Average Sugar Level by Manufacturer')
```

## IV. Statsmodels

```python
import statsmodels.api as sm
```

```python
# define X variables and y target variable
X = df_nutrition
y = df_cereal['rating'].values.reshape(-1,1)
```

```python
# fit ordinary least squares model and print results
reg = sm.OLS(y, X).fit()
print(reg.summary())
```
